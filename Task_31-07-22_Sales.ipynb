```python
#6 . try to find out in how many regions we are selling a product and who is a manager for a respective region 

#Ans in next line

manager = []
for row in df_merge['Region']:
    if row == 'Central':
        manager.append('Chris')
    elif row == 'East':
        manager.append('Erin')
    elif row == 'South':
        manager.append('Sam')
    elif row == 'West':
        manager.append('William')
    else:
        manager.append('NO MANAGER')
        
df_merge['manager'] = manager
```

```python
df_merge.groupby(['manager','Region'])['Row ID'].count().reset_index(name = 'Total Sales')
```

```python
#7 . find out how many different differnet shipement mode that we have and what is a percentage 
#usablity of all the shipment mode with respect to dataset 

#Ans
df_merge['Ship Mode'].unique() # 3 shipment mode we have

```

```python
d1 = df_merge.groupby(['Ship Mode'])['Row ID'].count().reset_index(name = 'Sales Count')
#df_merge.groupby(['Ship Mode'])['Row ID'].count()
df_ship_mode = pd.DataFrame(d1)
```

```python
df_ship_mode['Ship_Mode_Percentage']  = 100* df_ship_mode['Sales Count']/df_ship_mode.groupby(['Ship Mode'])['Sales Count'].transform('sum')
```
