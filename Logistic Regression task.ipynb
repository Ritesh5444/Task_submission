```python
# Scaling of the data

from sklearn.preprocessing import StandardScaler
scaler  = StandardScaler()
```

```python
X = X.drop('Region', axis = 1)
```

```python
#Scaling of data for the independent features
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#Chekcing of multicolinearity using VIF(Variance Inflation Factor)
vif = pd.DataFrame()
#print(vif)
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif['Features'] = X.columns
#vif['Features']
vif['Vif'] = [variance_inflation_factor(X_scaled,i) for i in range (X_scaled.shape[1])]
vif
```

Observation - We can see that the VIF value is high for DMC, DC, ISI, BUI, FWI


```python
#Splitting the dataset into train and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.25, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
#Model training

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg
```

```python
log_reg.fit(X_train,y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

#Observation - We can see that we are having 96 % accuracy with this model.

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test, y_pred)
conf_mat
```

```python
true_positive = conf_mat[0][0]
false_positive = conf_mat[0][1]
false_negative = conf_mat[1][0]
true_negative = conf_mat[1][1]
```

```python
#Accuracy

accuracy = (true_positive+ true_negative)/(true_positive+false_positive+true_negative+false_negative)
accuracy
```

```python
#Precision

precision = true_positive/(true_positive+false_positive)
precision
```

```python
#Recall

recall = true_positive/(true_positive+false_negative)
recall
```

```python
# F1 score

F1_score = 2*(recall*precision)/(recall + precision)
F1_score
```

```python
#Area under curve

from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test,y_pred)
auc
```
